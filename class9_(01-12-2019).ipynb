```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
'data2': range(3)})
pd.merge(df1, df2)
```

```python
# By default merge do inner join
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
'data2': range(3)})
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer') # will return all the values from both data
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
'data2': range(5)})
```

```python
 pd.merge(df1, df2, on='key', how='left') # will return all the values of left dataframe and NaN for missing values in right dataframe
```

```python
 pd.merge(df1, df2, on='key', how='right') # will return all the values of right dataframe and NaN for missing values in left dataframe
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
index=['a', 'c', 'e'],
columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
index=['b', 'c', 'd', 'e'],
columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
left2.join(right2, how='outer') # By default it will do left outer join
```

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1.join(right1, on='key') # left join
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```
